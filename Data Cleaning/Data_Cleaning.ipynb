```python
import pandas as pd
import numpy as np
df = pd.read_excel("Test Correlazione.xlsx", sheet_name=None)
df["Dati ETH"] = df["Dati ETH"].reset_index()
df["Dati ETH"].head()
```

```python
df["Dati Trading View"] = df["Dati Trading View"].reset_index()
df["Dati Trading View"].head()
```

```python
df["Dati ETH"]["open_time"] = pd.to_datetime(df["Dati ETH"]["open_time"])
df["Dati ETH"] = df["Dati ETH"].sort_values("open_time")
df["Dati Trading View"] = df["Dati Trading View"].sort_values("Date/Time")
df["Dati ETH"].rename(columns={"open_time": "DateTime"}, inplace=True)
df["Dati Trading View"].rename(columns={"Date/Time": "DateTime"}, inplace=True)
```

```python
df_paired = pd.merge_asof(
    df["Dati ETH"],
    df["Dati Trading View"],
    on = "DateTime",
    direction= "nearest"
)
```
